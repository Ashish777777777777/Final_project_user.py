```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

class RestaurantApp:
    def __init__(self):
        self.users = []
        self.food_items = [
            "Tandoori Chicken (4 pieces) [INR 240]",
            "Vegan Burger (1 Piece) [INR 320]",
            "Truffle Cake (500gm) [INR 900]"
        ]

    def register(self, full_name, phone_number, email, address, password):
        new_user = User(full_name, phone_number, email, address, password)
        self.users.append(new_user)
        print("Registration successful.")

    def login(self, email, password):
        for user in self.users:
            if user.email == email and user.password == password:
                return user
        return None

    def place_new_order(self, user):
        print("Available Food Items:")
        for index, food_item in enumerate(self.food_items, start=1):
            print(f"{index}. {food_item}")
        
        selected_indices = input("Enter the numbers of food items to order (separated by spaces): ")
        selected_indices = [int(idx) for idx in selected_indices.split()]
        
        selected_items = [self.food_items[idx - 1] for idx in selected_indices]
        
        print("Selected Items:")
        for item in selected_items:
            print(item)
        
        place_order = input("Do you want to place the order? (yes/no): ")
        if place_order.lower() == "yes":
            user.orders.append(selected_items)
            print("Order placed successfully.")
        else:
            print("Order not placed.")

    def order_history(self, user):
        print("Order History:")
        for idx, order in enumerate(user.orders, start=1):
            print(f"Order {idx}:")
            for item in order:
                print(item)
            print("-----------------------")

    def update_profile(self, user):
        print("Current Profile:")
        print(f"Full Name: {user.full_name}")
        print(f"Phone Number: {user.phone_number}")
        print(f"Email: {user.email}")
        print(f"Address: {user.address}")
        
        new_address = input("Enter new address: ")
        user.address = new_address
        
        print("Profile updated successfully.")

# Creating the application and performing operations
app = RestaurantApp()

app.register("John Doe", "1234567890", "john@example.com", "123 Main St", "password123")

logged_in_user = app.login("john@example.com", "password123")

if logged_in_user:
    print(f"Logged in as {logged_in_user.full_name}")
    
    while True:
        print("Options:")
        print("1. Place New Order")
        print("2. Order History")
        print("3. Update Profile")
        choice = input("Enter your choice: ")
        
        if choice == "1":
            app.place_new_order(logged_in_user)
        elif choice == "2":
            app.order_history(logged_in_user)
        elif choice == "3":
            app.update_profile(logged_in_user)
        else:
            print("Invalid choice. Exiting...")
            break
else:
    print("Login failed. Invalid email or password.")
```
